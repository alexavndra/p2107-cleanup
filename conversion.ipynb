# Conversion of mega dataframe + separation

## Importing dataframe + creating named columns for later use

```python
#! /usr/bin/python3
import os
import urllib.parse as ulp
import pandas as pd
import numpy as np
# import cgi
#import cgitb
#cgitb.enable(display=0, logdir='/var/www/cgi-bin/'i)

original_df = pd.read_csv("CCE-P2107_FINAL_EVENTLOG_20210813_190609.csv")

# For renaming
lon_lat = ["LonDec","LatDec"]
seasor_cols = ["LonDec","LatDec","Size","Angle","Font","Justify","Haul"]

# Columns to drop after
drop_cols = ['Message ID', 'Date', 'Event', 'Instrument', 'Action', 'Transect',
       'Station', 'Seafloor', 'Author', 'Comment', 'Cruise', 'R2R_Event', 'dateTimeUTC',
       'GPS_Time', 'dateTime8601', 'Revisions']

original_df
# original_df.shape (812 rows, 19 cols)
# original_df.columns
```

## Creating dataframes

```python
# Ensuring that the empty dataframes have columns
default_cols = ['Message ID', 'Date', 'Event', 'Instrument', 'Action', 'Transect',
       'Station', 'Cast', 'Latitude', 'Longitude', 'Seafloor', 'Author',
       'Comment', 'Cruise', 'R2R_Event', 'dateTimeUTC', 'GPS_Time',
       'dateTime8601', 'Revisions']

ctd_df = pd.DataFrame(columns=default_cols)

mocness_df = pd.DataFrame(columns=default_cols)

mvp_df = pd.DataFrame(columns=default_cols)

seasoar_df = pd.DataFrame(columns=default_cols)
```

## Importing data to correct frames

```python
for i in range(len(original_df)):
    if("ctd" in original_df.iloc[i,3].lower()):
        ctd_df.loc[len(ctd_df.index)] = original_df.loc[i]
        # df.loc[len(df.index)] = ['Amy', 89, 93]
    elif("mocness" in original_df.iloc[i,3].lower()):
        mocness_df.loc[len(mocness_df.index)] = original_df.loc[i]
    elif("mvp" in original_df.iloc[i,3].lower()):
        mvp_df.loc[len(mvp_df.index)] = original_df.loc[i]
    elif("seasoar" in original_df.iloc[i,3].lower()):
        seasoar_df.loc[len(seasoar_df.index)] = original_df.loc[i]
    else:
        continue
```

## Editing dataframes to have the correct columns + labels

### Fixing `ctd_df` dataframe

#### Dropping unnecessary columns

```python
ctd_df.drop(drop_cols, axis = 1, inplace = True)
ctd_df
```

#### Renaming the columns using `lon_lat`

```python
ctd_df = ctd_df.reindex(columns=["Longitude", "Latitude", "Cast"])
ctd_df.columns = ["LonDec","LatDec","Cast"]
ctd_df
```

### Fixing `mocness_df` dataframe

#### Dropping unnecessary columns

```python
mocness_df.drop(drop_cols, axis = 1, inplace = True)
mocness_df
```

#### Renaming columns + adjusting order

```python
mocness_df = mocness_df.reindex(columns=["Longitude", "Latitude", "Cast"])
mocness_df.columns = ["LonDec","LatDec","Haul"]
mocness_df_new_columns = ["LonDec","LatDec","Size","Angle","Font","Justify","Haul"]
mocness_df = mocness_df.reindex(columns = mocness_df_new_columns)
mocness_df
```
